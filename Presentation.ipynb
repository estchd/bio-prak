# Methods

## Data Files
- preprocessing:
    - original -> filter by chromosome 8
    - filtered -> rename all to chr8
    - 
- Assemble Regions with script
- Find Sequences for Regions with bedtools getfasta
- Intersect modifications with Regions with bedtools intersect
- Make modifications local to regions
- Combine modifications to single file region:modifications
- Calculate modified and unmodified accessibilities for regions
    - Window Sizes: [150, 400]
    - Footprints: [5, 30]
    - Started by using single threaded python and csv output
    - Switched to multi threaded rust and binary output for time and storage reasons 

## Modification Statistics
- number of modifications in different regions (exons vs introns)
    - count intersections
- Distance between modifications
    - Sort modifications in region
    - Compute pairwise distances both absolute and relative to sequence length
- Position of modifications in regions
    - Compute position of modification in region both absolute and relative to sequence length 

## Sanity Checking
- Check that every local modification is an A
    - First all at once and later during accessibility computations  
- Check that only looking at 2*windowsize slice doesnt introduce errors
    - Select random modification position where the full slice is in the region
    - Calculate accessibilities for slice and full sequence
    - Compare unmodified, modified and difference over slice window

## DRACH modifications
- Un-/Modified motifs
    - Go through whole chromosome sequence in 5 Basepair windows
    - Identify DRACH and RRACH motifs
    - Check if modification exists on position
- Modifications inside / outside motifs
    - Go through all modifications
    - Check if modification is inside DRACH or RRACH motif

## Aligned Accessibilities
- Get all exon regions
- Filter for single modification sites that have at least 2*window_size free space in front and back
- Calculate accessibilities for all filtered slices
- Calculate average difference in Probabilities over all computed accessibilities

## snoRNA modifications
- Identify snoRNA transcripts
- Check which snoRNA transcripts have modifications
- Compute folds for unmodified and modified transcripts through rnafold web server

## Regions with high accessibility changes
- Go through all exon region accessibilities
- Identify n region slices with maximum / minimum accessibility changes for every feature
- Get sequence for each region slice
- Fold sequence with and without modifications through rnafold web server

# Outlook
- What would you do differently?
    - Start out using Rust
        - Use type system to avoid off-by-one errors when indexing
        - Use type system to avoid field errors when deserializing 
        - Use Binary format for saving computed accessibility data for smaller file size and faster processing
        - Heavier and earlier use of multithreading to speed up computations and indexing
    - Write more results to file instead of variables
        - prevent slow recomputes because of needing to reset jupyter kernel
        - save file hashes from the start to prevent incomplete copies
    - Ask research questions and answer them instead of blind exploration of the data
