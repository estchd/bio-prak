# Milestone 1
Examination of the association of m6A sites with selected features

Statistical analyses of the distribution of m6A sites in the context of genomic and genic features, such as pre-RNA splicing, transcripts, transcript variants and/or sequence motifs, such as the DRACH motif. 

# Milestone 2
Computationally efficient evaluation of m6A sites and their potential to induce structural changes.

We suggest to use the "accessibility" measure presented in the course.
If you plan to use a different approach, please discuss it with the tutors/lecturers beforehand.

# Milestone 3
RNA switches: Effect of m6A on RNA secondary structure

Folding of promissing candidate RNA regions and visual representation of the effect of m6A on RNA secondary structure. We suggest to visualize a few example RNA structures with and without m6A from the set of your best candidates from Milestone 2 and interpret your results.
